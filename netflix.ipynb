#### What is the Correlation between the features?

```python
import pandas as pd

netflix_copy["date_added"] = pd.to_datetime(netflix_copy['date_added'], errors='coerce')
netflix_copy.dropna(subset=['date_added'], inplace=True)  # Drop rows with invalid dates
netflix_copy.head(2)
```

```python
netflix_copy['year_added'] = netflix_copy["date_added"].dt.year
netflix_copy['month_added'] = netflix_copy["date_added"].dt.month
netflix_copy.head(2)
```

```python
netflix_copy.dtypes
```

```python
import pandas as pd

# Select only numeric columns for correlation
numeric_cols = netflix_copy.select_dtypes(include=[np.number])
corr_matrix = numeric_cols.corr()

# Optional: You can explore correlations of non-numeric columns separately
# categorical_cols = netflix_copy.select_dtypes(exclude=[np.number])
# ... (further analysis on categorical columns)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'feature1': [1, 2, 3, 4, 5],
        'feature2': [6, 5, 4, 3, 2],
        'feature3': [4, 6, 5, 2, 1]}
netflix_copy = pd.DataFrame(data)

# Calculate correlation coefficients
correlation = netflix_copy.corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
sns.heatmap(correlation, annot=True)
plt.show()
```

##### Insight : Above Heatmap shows correlation between release_year,year_added & month_added.
